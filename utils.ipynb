```python

srl_predictor = load_predictor('structured-prediction-srl')
bert_predictor = load_predictor('structured-prediction-srl-bert')
```

```python
#TESTS FOR SRL MODEL

srl_output = []

def srl_data1(): 

  slists1 = []
  outlists1 = []

  for d in data1:
    output = srl_predictor.predict(d)
    out_1 = output['verbs'][0]['tags']

    slists1.append(d)
    outlists1.append(out_1)
          
    list_zip = zip(slists1, outlists1)
    zipped_list = list(list_zip)


    counter_x = 0
    counter_f = 0
    fails = []

    for x in zipped_list:
      counter_x += 1
      pred = x[1]
      if pred[-2].endswith('ARG2'):
        continue
      else:
        fails.append(x)
        counter_f += 1
          
  fail = (counter_f / counter_x) * 100

  srl_output.append(x)

  print('SRL FAILURE RATE- TEST 1:', fail, '%')
  print('SRL FAILED SENTENCES - TEST 1:', fails)

def srl_data2():

  slists2 = []
  outlists2 = []

  for d in data2:

    output = srl_predictor.predict(d)
    out_2 = output['verbs'][0]['tags']

    slists2.append(d)
    outlists2.append(out_2)


    list_zip = zip(slists2, outlists2)
    zipped_list = list(list_zip)


    fails = []
    counter_x = 0
    counter_f = 0

    for x in zipped_list:
      counter_x += 1
      pred = x[1]
      if pred[-2].endswith('TMP'):
        continue
      else:
        fails.append(x)
        counter_f += 1

      srl_output.append(x)

  fail = (counter_f / counter_x) * 100

  

  print('SRL FAILURE RATE - TEST 2:', fail, '%')
  print('SRL FAILED SENTENCES - TEST 2:', fails)

def srl_data3():


  slists3 = []
  outlists3 = []

  for d in data3:
    output = srl_predictor.predict(d)
    out_3 = output['verbs'][1]['tags']

    slists3.append(d)
    outlists3.append(out_3)

    list_zip = zip(slists3, outlists3)
    zipped_list = list(list_zip)


    fails = []
    counter_x = 0
    counter_f = 0


    for x in zipped_list:
      counter_x += 1
      pred = x[1]
      if pred[-2].endswith('ARG1'):
        continue
      else:
        fails.append(x)
        counter_f += 1

    srl_output.append(x)

  fail = (counter_f / counter_x) * 100

  

  print('SRL FAILURE RATE - TEST 3:', fail, '%')
  print('SRL FAILED SENTENCES - TEST 3:', fails)

def srl_data4():

  slists4 = []
  outlists4 = []

  for d in data4:
    output = srl_predictor.predict(d)
    out_4 = output['verbs'][1]['tags']

    slists4.append(d)
    outlists4.append(out_4)


    list_zip = zip(slists4, outlists4)
    zipped_list = list(list_zip)


    fails = []
    counter_x = 0
    counter_f = 0


    for x in zipped_list:
      counter_x += 1
      pred = x[1]
      if pred[-3].endswith('ARG1'):
        continue
      else:
        fails.append(x)
        counter_f += 1
    srl_output.append(x)

  fail = (counter_f / counter_x) * 100

  

  print('SRL FAILURE RATE - TEST 4:', fail, '%')
  print('SRL FAILED SENTENCES - TEST 4:', fails)

def srl_data5():
 

  slists5 = []
  outlists5 = []

  for d in data5:
    output = srl_predictor.predict(d)
    out_5 = output['verbs'][3]['tags']

    slists5.append(d)
    outlists5.append(out_5)


    list_zip = zip(slists5, outlists5)
    zipped_list = list(list_zip)


    fails = []
    counter_x = 0
    counter_f = 0


    for x in zipped_list:
      counter_x += 1
      pred = x[1]
      if pred[-7].endswith('ARG1'):
        continue
      else:
        fails.append(x)
        counter_f += 1

    srl_output.append(x)

  fail = (counter_f / counter_x) * 100

  print('SRL FAILURE RATE - TEST 5:', fail, '%')
  print('SRL FAILED SENTENCES - TEST 5:', fails)


def srl_data6():

  slists6 = []
  outlists6 = []

  for d in data6:
    output = srl_predictor.predict(d)
    out_6 = output['verbs'][1]['tags']

    slists6.append(d)
    outlists6.append(out_6)
  

    list_zip = zip(slists6, outlists6)
    zipped_list = list(list_zip)


    counter_x = 0
    counter_f = 0

    for x in zipped_list:
      counter_x +=1
      pred = x[1]
      if pred[3].endswith('ARG0'):
        continue
      else:
        counter_f += 1

    srl_output.append(x)

  fail = (counter_f / counter_x) * 100


  print('SRL FAILURE RATE - TEST 6:', fail, '%')
  print('SRL FAILED SENTENCES - TEST 6:', x)

def res_srl():
  with open('srl_output.txt', 'w', newline="") as x:
   csv.writer(x,delimiter=" ").writerows(srl_output)



  
```
